```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```
titanic_df = pd.read_csv(filepath_or_buffer="https://raw.githubusercontent.com/learningwithkamal/ML_Basics/master/Titanic_train.csv")
titanic_df = titanic_df.drop(['Age_wiki'], axis=1)
print (titanic_df.shape)
titanic_df.head()
```

# **Seaborn Catplot**

```
sns.catplot(x="Sex", y="Age", data=titanic_df)
```

```
sns.catplot(x="Sex", y="Age", jitter=False, data=titanic_df)
```

```
sns.catplot(x="Sex", y="Age",  kind="swarm",data=titanic_df)
```

```
sns.catplot(x="Sex", y="Age", hue="Survived" ,kind="swarm",data=titanic_df)
```

```
sns.catplot(x="Pclass", y="Sex", hue="Survived" ,kind="swarm",data=titanic_df)
```

# **Box Plot**

```
sns.catplot(x="Survived", y="Age", kind="box", data=titanic_df)
```

```
sns.catplot(x="Survived", y="Age", kind="box",hue="Sex", data=titanic_df)
```

```

sns.catplot(x="Survived", y="Age", kind="violin",hue="Class",  data=titanic_df)
```

```
sns.catplot(x="Survived", y="Age", kind="violin",hue="Sex",spilt=True , data=titanic_df)
```

# **Bar PLot**

```
sns.catplot(x="Sex", y="Survived", hue="Class", kind="bar", data=titanic_df);
```

```
g = sns.catplot(x="Fare", y="Survived", row="Class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic_df.query("Fare > 0"))
g.set(xscale="log");
```

## **Seaborn Count Plot**

```

sns.countplot(x = 'Pclass', hue = 'Survived', palette = 'Set1', data = titanic_df)
```

```
sns.countplot(x = 'Sex', hue = 'Survived', palette = 'Set1', data = titanic_df)
```

# **Seaborn Factor Plot**

```
sns.catplot(x="Sex", y="Survived", col="Pclass",
                    data=titanic_df, saturation=.5,
                    kind="bar", ci=None, aspect=.6)
```

**Correlation betwe**en **features**

```
titanic_df.corr()["Survived"]
```

```
sns.regplot(x="Survived", y="Class", data=titanic_df);
```

# **Visualizing the distribution of a dataset**

# Plotting Univariate Distribution using Seaborn KDEPlot

```
sns.kdeplot(titanic_df['Age'], shade=True, cut=0) 
```

# Fitting Parametric Distribution using Seaborn DistPlot




```
from scipy import stats

sns.distplot(titanic_df['Age'], kde=True, fit=stats.gamma);
```

## Plotting Bivariate distributions using Seaborn JointPlot





```
sns.jointplot(x=titanic_df['Age'], y=titanic_df['Pclass'], data=titanic_df);
```

## **# Visualizing pairwise relationships in a dataset using Seaborn PairPlot**

```
sns.pairplot(titanic_df,hue="Survived")
```

```
sns.pairplot(titanic_df[['Class','Age','Sex','Survived','Pclass']])
```
